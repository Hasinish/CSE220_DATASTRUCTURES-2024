```python
class Node_pair:
  def __init__(self, key, value, next = None):
    self.key, self.value, self.next = key, value, next


class Hashtable:
  def __init__(self, size):
    self.ht = [None]*size


  def insert(self, s):
    hashed_index = self.__hash_function(s[0])
    pair = Node_pair(s[0], s[1])
    if self.ht[hashed_index] == None:
      self.ht[hashed_index] = pair
    else:
      pair.next = self.ht[hashed_index]
      self.ht[hashed_index] = pair


  def create_from_array(self, arr):
    for i in arr:
      self.insert(i)

  def print_hashtable(self):
    idx = 0
    for i in self.ht:
      print(idx, ':', end = ' ')
      head = i
      while head != None:
        print(f'({head.key}, {head.value})', end = '-->')
        head = head.next
      print('None')
      idx += 1


  #Do it by yourself
  def __hash_function(self, key):
    #TO DO
    sum = key + 3
    return sum % 6


  #Do it by yourself
  def remove(self, key):
    #TO DO
    idx = self.__hash_function(key)
    new = self.ht[idx]

    if new == None :
      return self.ht

    elif new.key == key :

      if new.next != None:
        self.ht[idx] = new.next

      else:
        self.ht[idx] = None

    elif new.key != key and new.next != None :

      new = new.next
      while new != None :

        if new.key == key:
          if new.next != None :
            self.ht[idx].next = new.next

          else :
            self.ht[idx].next = None
        new = new.next

    return self.ht




```
